## Read in the data

```python
# import the data 
import pandas as pd

df = pd.read_csv('happiness_survey_data.csv')
df.head()
```

```python
# there are 2089 rows and 6 columns of data
df.shape
```

```python
# the years range from 2005 - 2021
# happiness scores range from 2 - 8
df.describe()
```

## Explore the data

```python
# Filter out any data before 2010 and after 2019
mask = (df['year'] >= 2010) & (df['year'] <= 2019)
filtered_df = df.loc[mask]

filtered_df
```

```python
# Group the data by country and calculate the maximum happiness score for each one
country_max_df = df.groupby('country_name')['happiness_score'].max().reset_index()
country_max_df.head()
```

```python
# Sort the grouped countries by happiness score and return the top five
country_max_df.sort_values('happiness_score', ascending=False)
country_max_df.head()
```

```python
# Group the data by country and calculate the average happiness score for each one
country_mean_df = df.groupby('country_name')['happiness_score'].mean().reset_index()
country_mean_df.head()
```

```python
# Sort the grouped countries by happiness score and return the top five
country_mean_df.sort_values('happiness_score', ascending=False)
country_mean_df.head()
```

# Prepping for Modeling

## 1. Create a Single Table

Hi there,

We just finished collecting our Q2 book sales data. Can you help us create one giant table that includes:

* April, May and June’s book sales
* Customer data

Each table has a field called customer_id that links all the tables together.

I’ve attached four files.

Thanks!\
Brooke

```python
# Read all four files into a Jupyter Notebook
import pandas as pd

april = pd.read_excel('Book_Sales_April.xlsx')
may = pd.read_excel('Book_Sales_May.xlsx')
june = pd.read_excel('Book_Sales_June.xlsx')
customers = pd.read_csv('Book_Customers.csv')
```

```python
# Concatenate the April, May and June book sales DataFrames
sales = pd.concat([april, may, june])
sales.head()
```

```python
# Join the newly created book sales DataFrame with the customers DataFrame on customer_id
# Which type of join do you think would work best here?
df = sales.merge(customers, how = 'right', on='Customer ID')
df.head()
```

## 2. Prepare Rows for Modeling

Hi again,

We are trying to predict which customers will purchase a book this month.

Can you reformat the data you compiled earlier this week so that it’s ready to be input into a model, with each row representing a customer instead of a purchase?

Thanks!\
Brooke

```python
# What should one row of data look like for the model?
## customer - purchased in June or not - summary stats about April and May
```

```python
# Create a column called 'June Purchases' that sums up all purchases in June
june_data = df[df['Purchase Date'].dt.month == 6]
june_purchases = june_data.groupby('Customer ID').size().reset_index(name='June Purchases')
june_purchases.head()
```

```python
# Only look at April and May data
df_april_may = df[df['Purchase Date'].dt.month < 6]
df_april_may.head()
```

```python
# Create a column called 'Total Spend' that sums up the prices of all the books that were purchased in April and May
total_spend = df_april_may.groupby('Customer ID')['Price'].sum().reset_index(name='Total Spend')
total_spend.head()
```

```python
# Combine the 'June Purchases' and 'Total Spend' columns into a single DataFrame for modeling
model_df = total_spend.merge(june_purchases, how='left', on='Customer ID').fillna(0)
model_df.head()
```

## 3. Prepare Columns for Modeling

Hi again,

Thanks for your help earlier this week! We just learned that we also have to make all of the data numeric before inputting it into a predictive model.

Can you turn the Audience text field into a numeric field?

Thanks!\
Brooke

```python
df_april_may.head()
```

```python
# Change the Audience field into a dummy variable
audience_dummies = pd.get_dummies(df_april_may['Audience'])
audience_dummies.head()
```

```python
# Combine back with the customer table
cust_audience = pd.concat([df_april_may['Customer ID'], audience_dummies], axis=1)
cust_audience.head()
```

```python
# Using the Audience dummy variables, create three new columns that contain
# the number of Adult / Children / Teen books purchased by each customer
book_types = cust_audience.groupby('Customer ID').sum().reset_index()
book_types.head()
```

```python
# Combine the three new columns back with the customer-level data
model_df = model_df.merge(book_types, how = 'left', on = 'Customer ID')
model_df.head()
```

## 4. Feature Engineering

Hi again,

I have one final request for you.

As a reminder, our goal is to try and predict which customer will purchase a book this month.

Can you create new features that you think will do a good job making a prediction?

Thanks!\
Brooke

```python
# our original dataframe
df.head()
```

```python
# dataframe to be input into a model - single table, non-null, numeric with a y column and many x columns
model_df.head()
```

```python
# what features could do a good job predicting june purchases?
## whether a customer has school-aged children
## percent of the time that a customer purchases online vs in person
customers.head()
```

```python
# create the feature
# whether a customer has school-aged children 
customers['Has School Aged Children Numeric'] = customers['Has School Aged Children'].map({'Yes': 1, 'No': 0})
has_kids = customers[['Customer ID', 'Has School Aged Children Numeric']]
has_kids.head()
```

```python
# add it to the model dataframe
model_df = model_df.merge(has_kids, how='left', on='Customer ID')
model_df.head()
```

```python
df_april_may.head()
```

```python
# create another feature
# make the purchase location numeric
location = pd.get_dummies(df_april_may['Purchase Location'])
```

```python
# add it on to the original dataframe
location_df = pd.concat([df_april_may, location], axis=1)[['Customer ID', 'In Person', 'Online']]
location_df.head()
```

```python
# group the locations by customer
location_group = location_df.groupby('Customer ID').sum().reset_index()
location_group.head()
```

```python
# calculate the percent in person purchases by customer
in_person_pct = (location_group['In Person'] / (location_group['In Person'] + location_group['Online'])*100).rename('Percent In Person')
in_person_pct.head()
```

```python
# add it on to the model dataframe
model_df = pd.concat([model_df, in_person_pct], axis=1)
model_df.head()
```

## 5. The Final DataFrame for Modeling

```python
# our non-null, numeric dataframe with y and x variables
model_df
```

```python
# y = June Purchases
# x's = Total Spend through Percent in Person
# would need to remove Customer ID for modeling
```

```python
# the final dataframe that is ready to be input into a model!
# Use the correct column name as it appears in model_df
final_df = model_df[['June Purchases', 'Total Spend', 'Adults', 
                     'Children', 'Teens', 'Has School Aged Children Numeric', 'Percent In Person']]

final_df = final_df.rename(columns={'Has School Aged Children Numeric': 'Has Kids'})
final_df = final_df.fillna(0)  # Or you can use .dropna() if preferred
final_df.head()
```
