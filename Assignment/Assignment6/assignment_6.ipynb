### Model Fitting

```python
# fitting model on `ram` and constant, with 'price' as target

import statsmodels.api as sm

X = sm.add_constant(computers["ram"])
y = computers["price"]

model = sm.OLS(y,X).fit()

model.summary()
```

### Intepretation:

A one megabyte increase in RAM is associated with a $64.29 increase in price.

```python
feature_values = [0, 2, 4, 8, 16, 32, 64]

predict_df = sm.add_constant(pd.DataFrame({"ram": feature_values}))

predict_df
```

```python
# pass new feature dataframe into model

predictions = model.predict(predict_df)

predictions.index = feature_values

predictions
```

### Model Fitting & Evaluation

```python
# Independence - DW Test Statistic looks great

cols = ["ram", "speed", "hd", "screen", "ads", "trend"]

X = sm.add_constant(computers.loc[:, cols])
y = computers["price"]

model = sm.OLS(y, X).fit()

model.summary()
```

```python
# looks like slight equal variance issues

sns.scatterplot(x= model.fittedvalues, y= computers["price"])

sns.despine()
```

```python
# Normality is violated - try a log transform

import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(model.resid, dist="norm", plot=plt);
```

```python
### Residual plotter helper function

def residual_analysis_plots(model):
    
    import scipy.stats as stats
    import matplotlib.pyplot as plt
    
    predictions = model.predict()
    residuals = model.resid
    
    fig, ax = plt.subplots(1, 2, sharey="all", figsize=(10, 6))
    
    sns.scatterplot(x=predictions, y=residuals, ax=ax[0], alpha=.3)
    ax[0].set_title("Residual Plot")
    ax[0].set_xlabel("Prediction")
    ax[0].set_ylabel("Residuals")
    
    stats.probplot(residuals, dist="norm", plot=ax[1])
    ax[1].set_title("Normal Q-Q Plot")
```

```python
residual_analysis_plots(model)
```

# 4. Data Splitting

1. Split your data into a training and test set
2. Use cross validation to fit a model on all numeric features. Report r2 values for each validation fold.
3. Fit your model on all of your training data and score on the test dataset. 

### Test Data Split

```python
from sklearn.model_selection import train_test_split


features = ["speed", "hd", "ram", "screen", "ads", "trend"]

X = sm.add_constant(computers[features])
y = computers["price"]

# Test Split (test_size=.2, random_state=12345)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state = 12345)

print(
    f"Training Set Rows: {X.shape[0]}",
    f"Test Set Rows: {X_test.shape[0]}",
)
                    
```

### Cross Validation Loop

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_absolute_error as mae

kf = KFold(n_splits=5, shuffle=True, random_state=2023)

val_scores = []

for train_ind, val_ind in kf.split(X, y):
    X_train, y_train = X.iloc[train_ind], y.iloc[train_ind]
    X_val, y_val = X.iloc[val_ind], y.iloc[val_ind]

    model = sm.OLS(y_train, X_train).fit()
    val_scores.append(r2(y_val, model.predict(X_val),))

print("All Validation Scores:", [round(x, 3) for x in val_scores])
print(f"Cross Val Score: {round(np.mean(val_scores), 3)} +- {round(np.std(val_scores), 3)}")
```

### Model Fit on All Training Data

```python
model = sm.OLS(y, X).fit()

model.summary()
```

### Score on Test Data

```python
y_test_pred = model.predict(X_test)  # Predictions for the test set

print(f"Test R2: {r2(y_test, y_test_pred):.4f}")  
print(f"Test MAE: {mae(y_test, y_test_pred):.2f}")
```

Create the following features and determine if they improve model accuracy.

2. Then, try combining the 'cd' and 'multi' features: If 'cd' AND 'multi' both equal "yes" then 1, else 0.

```python
import pandas as pd

computers = pd.read_csv("Computers.csv")

computers.head()
```

### Feature Engineering

```python
ram_dict = {
    2: "<=4",
    4: "<=4",
    8: "8",
    16: "16+",
    24: "16+",
    32: "16+"    
}

computers_eng = (
    pd.get_dummies(
        computers.assign(
            hd2 = computers["hd"] ** 2,
            hd3 = computers["hd"] ** 3,
            cd_multi_combined=np.where(
                (computers["cd"] == "yes") & (computers["multi"] == "yes"), 1, 0
            ),            
            ram_categories=computers["ram"].map(ram_dict)  
        )
        .drop(["ram", "cd", "multi"], axis=1)
    ,drop_first=True    
    ).astype(int)
)

computers_eng.head()
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split


features = ["speed", "hd", "ram", "screen", "ads", "trend"]

X = sm.add_constant(computers_eng.drop("price", axis=1))
# X = sm.add_constant(computers[features])
y = np.log(computers["price"])

# Test Split (test_size=.2, random_state=12345)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12345)
```

### Cross-Validation

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_absolute_error as mae

kf = KFold(n_splits=5, shuffle=True, random_state=2023)
r2_scores = []
mae_scores = []

for train_idx, val_idx in kf.split(X_train):
    X_train_fold, X_val_fold = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]
    
    model = sm.OLS(y_train_fold, X_train_fold).fit()
    y_pred = model.predict(X_val_fold)
    
    r2_scores.append(r2(y_val_fold, y_pred))
    mae_scores.append(mae(y_val_fold, y_pred))

print(f"Cross-Validation R2 Scores: {r2_scores}")
print(f"Cross-Validation MAE Scores: {mae_scores}")
print(f"Average R2: {np.mean(r2_scores):.4f}")
print(f"Average MAE: {np.mean(mae_scores):.2f}")
```
